```python
# Import libraries
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.graphics.tsaplots import month_plot, quarter_plot, plot_acf, plot_pacf
from statsmodels.tsa.seasonal import seasonal_decompose
from sklearn.metrics import root_mean_squared_error, mean_absolute_error, mean_absolute_percentage_error
from statsmodels.tsa.holtwinters import ExponentialSmoothing
```

```python
# Set the index when we import the data
df = pd.read_csv("airmiles.csv",
                 index_col = "Date",
                 parse_dates=True,
                 dayfirst = True)
df.head()
```

```python
# Information about the df
df.info()
```

```python
# Change the name of the time series variable t y
df = df.rename(columns = {"airmiles": "y"})
df.head()
```

```python
# Rename the Y variable
df = df.rename(columns = {"airmiles": "y"})
df.head()
```

# Task 1

```python
# Set index to monthly frequency MS
```

```python
# Change the frequency
df = df.asfreq('MS')
df.index
```

# Data Visualization

```python
# Daily Closing Price Plot
df["y"].plot()
plt.show()
```

```python
# Simple time series plot
df.y.plot(figsize=(10, 4))
plt.show()
```

# Seasonality

```python
# Plot the month_plot
month_plot(df.y, ylabel = "Monthly Air Miles")
plt.show()
```

```python
# Plot the quarter_plot
quarter_plot(df.y.resample("QE").sum(), ylabel = "Quarter Air Miles")
plt.show()
```

```python
# Seasonal decomposition for df['Adj. Close']
decomposition = seasonal_decompose(df.y,
                                  model = 'mul',
                                  period = 12)
fig = decomposition.plot()
fig.set_size_inches(10, 8)
plt.show()
```

# (Partial) Auto-Correlation

```python
# Plot the ACF of the
plot_acf(df.y, lags = 40)
plt.show()
```

```python
# PACF
plot_pacf(df.y, lags=40)
plt.show()
```

# Model Assessment

```python
# Set the test set to be the last 12 months
df_train = df.iloc[:-12]
df_test = df.iloc[-12:]
```

```python
# Build the Holt-Winters Model
model = ExponentialSmoothing(
    df_train.y,
    trend = 'add',
    seasonal = 'mul',
    seasonal_periods = 12).fit()
```

# Plot the future

```python
# Predict 12 periods with the model
predictions = model.forecast(steps = len(df_test)).rename("Holt-Winters)")
predictions
```

```python
# Assess the model with the MAE, MAPE, and RMSE
mae = mean_absolute_error(df_test.y, predictions)
mape = mean_absolute_percentage_error(df_test.y, predictions)
rmse = root_mean_squared_error(df_test.y, predictions)

# Display the results
print(f"MAE: {mae:.2f}")
print(f"MAPE: {mape:.2%}")
print(f"RMSE: {rmse:.2f}")
```

```python
# Visualise the training, test and predcitions data
df_train.y.plot(figsize = (10, 4))
df_test.y.plot()
predictions.plot()
plt.show()

```

# Predict the future

```python
# build a holt winters model with the complete data set
## Fit the data
model_complete = ExponentialSmoothing(
    df.y,
    trend = 'add',
    seasonal = 'mul',
    seasonal_periods = 12).fit()

# Predict the next 12 months
future_predictions = model_complete.forecast(steps = 12).rename("Holt-Winters)")


# Visualise training and future predictions
df.y.plot(figsize = (10, 4))
future_predictions.plot()
plt.show()
```
